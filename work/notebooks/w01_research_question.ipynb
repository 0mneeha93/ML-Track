**Question:** What performance archetypes exist across the content inventory?

**Unit of analysis:** one content item (one page), each row represents a single page's behavior profile.

**The decision:** which broad action bucket a page belongs to protect, improve, rewrite, merge, prune, or monitor.

**The action:** a content team member reviews pages cluster by cluster rather than one at a time, applying a shared treatment to each archetype.

**Cost of a wrong call:** a misclustered page could get the wrong treatment e.g. spending effort improving a low demand page that isn't worth it, or leaving a genuinely valuable page unprotected because it was grouped with weaker pages.

**Scope note:** this dataset contains no article text, so this is structured/metric-based clustering not semantic clustering.

Across 30,000 pages, there's real structural diversity worth clustering: content type is heavily skewed (27,207 keyword articles vs. just 697 comparison articles), trend_direction spans five distinct behavioral states (from 16,262 declining pages to 1,152 flat ones), and word_count ranges enormously from 8 words to 9,546, with a median around 2,877. This spread across type, trend, and depth suggests pages aren't behaving as one uniform group, which is exactly the kind of variation structured clustering could organize into meaningful archetypes.

```python
import os, subprocess

REPO_URL = "https://github.com/0mneeha93/ML-Track"
REPO_DIR = "ML-Track"

if not os.path.isdir(REPO_DIR):
    subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
os.chdir(REPO_DIR)

print("Working dir:", os.getcwd())

import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total pages:", df.shape[0])
print("\n--- content_type distribution ---")
print(df["content_type"].value_counts())

print("\n--- trend_direction distribution ---")
print(df["trend_direction"].value_counts())

print("\n--- word_count spread ---")
print(df["word_count"].describe()[["min","25%","50%","75%","max"]])
```

This clustering is based on observable behavioral and metadata signals only not article text, so results describe structural/metric patterns, not semantic meaning. Cluster groupings reflect patterns observed in this anonymized starter slice; they are not proven causal categories.
