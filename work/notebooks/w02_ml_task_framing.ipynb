## 3. Success metric
**Silhouette score.** This measures how well separated clusters are it compares how close a page is to others in its own cluster versus the nearest different cluster, on a scale from -1 to 1. Higher means tighter, more distinct clusters. I'll defend "good" as: a silhouette score meaningfully above 0 (weak/overlapping clusters), combined with a manual sense-check, pulling a few real example pages from each cluster and confirming they genuinely resemble each other and look different from pages in other clusters. Silhouette score alone can be misleading (a mathematically tight cluster can still be meaningless in practice), so both checks matter together.

## 5. Why ML beats a fixed rule here
A fixed rule can separate pages cleanly on one or two dimensions for example, Notebook 02's "stale AND visible" rule. But archetypes plausibly depend on combinations across many signals at once: content type, word count, engagement, freshness, CTR, and trend, all interacting simultaneously. Writing an if-statement for every meaningful combination of 10 features would mean manually enumerating dozens of branches, and I wouldn't even know which combinations matter without first looking at the data. Clustering discovers these groupings automatically across all dimensions at once, surfacing patterns a human wouldn't think to hand-check individually. This matches the framing-ml problems skill's guidance: ML earns its place when signals are "many, tangled, and shifting" not when a single clean threshold would do.
This output feeds directly into the same content review action from w01: a reviewer treats each archetype as a group, protect, improve, rewrite, merge, prune, or monitor, rather than reviewing 30,000 pages individually.
